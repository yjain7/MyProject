```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
# Calculate silhouette_score
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")
```

```python
#Read files
trainfile = r'/Users/yashjain/Downloads/SouthGermanCredit.csv'
trainData = pd.read_csv(trainfile) #creates a dataframe


print(trainData.shape)
```

```python
#check the datatypes
trainData.info()
```

```python
#check for missing values

trainData.isnull().sum().sort_values(ascending=False)
```

```python
#preprocess data
#drop the target column 
X_train= trainData.drop(['kredit'], axis=1)
```

```python
#Scaling the dataset
sd = StandardScaler()
sd.fit(X_train)
scaled= pd.DataFrame(sd.transform(X_train),columns = X_train.columns)
scaled.describe()
```

```python
X_train.isnull().sum().sort_values(ascending=False)
```

```python
#K-Means Clustering ========================================================================
#Default Mode - K=8
kmeans = KMeans()
kmeans.fit(X_train)
print(kmeans.labels_)
print(kmeans.cluster_centers_)
#Print Silhouette measure
print(silhouette_score(X_train, kmeans.labels_))
print(X_train.shape)
```

```python
#hyperparameter tuning for k-mean clustering
# K=4
kmeans = KMeans(n_clusters = 4, init= 'random', algorithm= 'elkan')
kmeans.fit(X_train)
print(kmeans.labels_)
print(kmeans.cluster_centers_)
#Print Silhouette measure
print(silhouette_score(X_train, kmeans.labels_))
print(X_train.shape)
```

```python
#Add Cluster Number to each datapoint + save file
df_kmeans=X_train.copy()
kmeans_predict_train=kmeans.predict(X_train)
print(kmeans_predict_train)

df_kmeans['Cluster Number']=kmeans.labels_
print(X_train.shape)
print(df_kmeans.shape)
print(df_kmeans.head())
df_kmeans.dtypes

export_csv = df_kmeans.to_csv(r'/Users/yashjain/Downloads/KmeansClustersProb_2.csv')
```
